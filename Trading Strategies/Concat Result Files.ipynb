Concat Files

```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('Jan-May Data with Trading Signals.csv')
test_data = pd.read_csv('June Data with Trading Signals.csv')
combined_data = pd.concat([train_data, test_data], axis=0)
combined_data.reset_index(drop=True, inplace=True)
combined_data = combined_data.drop(columns=['Unnamed: 0'])
combined_data = combined_data.rename(columns={'Trading Signal': 'Signal'})

combined_data.head(20)

```

```python
combined_data.to_csv('All Trading Simulation Output.csv')
```
