```python
import pandas as pd

# Read CSV file
df = pd.read_csv('UoB_Set01_2025-01-02tapes.csv', header=None, names=['timestamp', 'price', 'number'])

# Handle duplicate timestamps
df_cleaned = df.groupby('timestamp').agg({'price': 'mean', 'number': 'sum'}).reset_index()

# Rename columns
df_cleaned = df_cleaned.rename(columns={'price': 'price', 'number': 'number'})

# Round to one decimal place
df_cleaned['price'] = df_cleaned['price'].round(1)

# Save cleaned data to a new file
df_cleaned.to_csv('cleaned_data.csv', index=False)

# Display the first few rows of the cleaned data
print(df_cleaned.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
df = pd.read_csv('cleaned_data.csv')

# Sample every 100th row
sampled_df = df.iloc[::100]

# Plot time series
plt.plot(sampled_df.index, sampled_df['price'], marker='o', label='Price')
plt.plot(sampled_df.index, sampled_df['number'], marker='s', label='Number')

# Add title and labels
plt.title('Time Series Plot (Sampled)')
plt.xlabel('Index')
plt.ylabel('Value')

# Add legend
plt.legend()

# Show plot
plt.show()
```
